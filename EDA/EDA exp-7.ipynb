```python
import geopandas as gpd
import requests
import zipfile
import os

# Function to download and extract ZIP files
def download_and_extract(url, extract_to='.'):
    response = requests.get(url)
    zip_path = os.path.join(extract_to, 'temp.zip')

    with open(zip_path, 'wb') as f:
        f.write(response.content)

    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_to)

    os.remove(zip_path)  # Clean up the temporary zip file

# Download and extract India states and districts
download_and_extract('https://gadm.org/data/shp/IND_adm1.zip', extract_to='content/data/india/')
download_and_extract('https://gadm.org/data/shp/IND_adm2.zip', extract_to='content/data/india/')

# Load the world map from downloaded files
world_map = gpd.read_file('path_to_downloaded/natural_earth_admin0.shp')  # Adjust this path based on your download location

# Load the extracted shapefiles for states and districts
india_states_map = gpd.read_file('data/india/IND_adm1.shp')
india_districts_map = gpd.read_file('data/india/IND_adm2.shp')

# Merge world data with the world map
world_data_geo = world_map.merge(world_data, how='left', left_on='name', right_on='Country')

# Merge Indian states and districts data
india_states_data_geo = india_states_map.merge(india_states_data, how='left', left_on='NAME_1', right_on='State')  # Adjust according to your state data
india_districts_data_geo = india_districts_map.merge(india_districts_data, how='left', left_on='NAME_2', right_on='District')  # Adjust according to your district data

```
