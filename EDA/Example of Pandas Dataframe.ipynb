```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
data={"Name":["ABC","EFG","HIJ","KLM"],"Age":np.random.randint(0,65,4),"Salary":np.random.randint(10000,1000000,4),"Occupation":["Worker","Employee","Staff","Teacher"]}
df=pd.DataFrame(data,columns=data.keys())
df
```

```python
print(df.describe())
print(df.head())

print(df.info())
print(df['Age'].mean())
print(df['Salary'].mean())
```

```python
plt.figure(figsize=(8,10))
plt.bar(df["Name"],df["Salary"])
plt.xlabel("Name")
plt.ylabel("Salary")
plt.title("Salary of Employees")
plt.show()
```
