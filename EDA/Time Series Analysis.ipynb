```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf
```

```python
# Data Preparation
np.random.seed(0)
date_range = pd.date_range(start='2020-01-01', end='2023-12-31', freq='M')
sales_data = np.random.randint(10000, 50000, size=len(date_range))
df = pd.DataFrame({'Sales': sales_data}, index=date_range)
```

```python
# Summary Statistics
print(df.describe())
```

```python
# Plot Sales Data
df['Sales'].plot(figsize=(12, 6), title='Monthly Sales Over Time', grid=True)
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

```python
# Decomposition
decomposition = seasonal_decompose(df['Sales'], model='additive', period=12)
decomposition.plot()
plt.show()
```

```python
# Autocorrelation Plot
plot_acf(df['Sales'], lags=24, title='Autocorrelation Plot')
plt.grid(True)
plt.show()
```

```python
# Rolling Mean and Standard Deviation
df['Sales'].plot(label='Sales', figsize=(12, 6))
df['Sales'].rolling(window=12).mean().plot(label='Rolling Mean (12 months)', color='red')
df['Sales'].rolling(window=12).std().plot(label='Rolling Std (12 months)', color='green')
plt.title('Rolling Mean and Standard Deviation')
plt.xlabel('Date')
plt.legend()
plt.grid(True)
plt.show()
```
